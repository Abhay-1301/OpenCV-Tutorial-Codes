```python
# Canny Edge Detection

import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('lena.jpg',cv2.IMREAD_GRAYSCALE)
canny = cv2.Canny(img,100,200) #add trackbar and see how edge detection changes when you change threshold 1 and threshold 2
#lesser noise in canny edge detection

titles = ['image','canny']
images = [img,canny]

for i in range(len(images)):
    plt.subplot(1,2,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
